```python
from pathlib import Path

def rename_level_only_folder(directory, dry_run=True):
    directory = Path(directory)

    for path in directory.iterdir():  # <-- NOT recursive
        name = path.name

        if path.is_file():
            stem = path.stem

            if stem.startswith("BC"):
                stem = stem[2:].lstrip(".")

            new_stem = stem.replace(".", "_") + "_IHC"
            new_name = new_stem + path.suffix

        elif path.is_dir():
            folder_name = name

            if folder_name.startswith("BC"):
                folder_name = folder_name[2:].lstrip(".")

            new_name = folder_name.replace(".", "_") + "_IHC"

        else:
            continue

        if new_name != name:
            print(f"{name} → {new_name}")
            if not dry_run:
                path.rename(path.with_name(new_name))



rename_level_only_folder(WSI_DIR,dry_run = False)
```

```python
#get all names files

paths = glob.glob(WSI_DIR +'*')
images_dict = {'complete_file_name':[], 'id':[],'stain_method':[]}
for path in paths:
    file_name = Path(path).name
    if file_name.endswith(EXT):
        file_stem = Path(path).stem 
        patient_id = file_stem.split("_")[0]
        stain_method = file_stem.split("_")[-1]
        images_dict['complete_file_name'].append(file_stem[:-4]+EXT)
        images_dict['id'].append(patient_id)
        images_dict['stain_method'].append(stain_method)

images_dict
```

```python
import pandas as pd
new_files_df = pd.DataFrame(images_dict)

new_files_df.to_csv(WSI_DIR+"chu_data_08012026.csv", index=False,header=True)
```

```python
#images dont besoin de label
FORMER_DIR =  "D:/data/images/bloc/"
paths = glob.glob(FORMER_DIR +'*')
images_dict = {'complete_file_name':[], 'id':[],'stain_method':[]}
for path in paths:
    file_name = Path(path).name
    if file_name.endswith(EXT):
        file_stem = Path(path).stem 
        patient_id = file_stem.split("_")[0]
        stain_method = file_stem.split("_")[-1]
        if patient_id not in images_dict['id']:
            images_dict['id'].append(patient_id)
            images_dict['complete_file_name'].append(file_stem[:-4]+EXT)
            images_dict['stain_method'].append(stain_method)

images_dict

```

```python
former_img_df = pd.DataFrame(images_dict)
former_img_df['id'].to_csv(FORMER_DIR+'chu_data_octobre2025_paired.csv',index = False, header=True)
```
